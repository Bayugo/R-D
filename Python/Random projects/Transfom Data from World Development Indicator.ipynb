```python
import pandas as pd

# Load your dataset
file_path = "Data.csv"  # Replace with your file's actual path
df = pd.read_csv(file_path)

# Melt the year columns into rows
df_melted = pd.melt(
    df,
    id_vars=["Country Name", "Country Code", "Series Name", "Series Code"],
    var_name="Year",
    value_name="Value"
)

# Clean up the 'Year' column (remove "[YRXXXX]")
df_melted["Year"] = df_melted["Year"].str.extract(r"(\d{4})")

# Save or inspect the transformed DataFrame
print(df_melted.head())
df_melted.to_csv("transformed_data.csv", index=False)
```

```python


# Load the data into a DataFrame (replace 'your_file.csv' with your file path)
df = pd.read_csv('transformed_data.csv')

# Reshape the DataFrame using pivot
df_pivot = df.pivot(index=['Country Name', 'Country Code', 'Year'], 
                    columns='Series Name', 
                    values='Value').reset_index()

# Rename columns to be more concise if needed
df_pivot.columns.name = None  # Remove multi-level column name if pivoted DataFrame has one

# Display the transformed DataFrame
print(df_pivot)

# Save the transformed data to a CSV file (optional)
df_pivot.to_csv('ultra_transformed_data.csv', index=False)
```
